```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
negative = r'Z:\Programming\MyML\NLP COLLAGE\labeling the data\NLP_Research_paper\negative.txt'
positive = r'Z:\Programming\MyML\NLP COLLAGE\labeling the data\NLP_Research_paper\Positive.txt'
neutral = r'Z:\Programming\MyML\NLP COLLAGE\labeling the data\NLP_Research_paper\Netural.txt'
```

```python
def sentiment_scores(positive):
    # Create a SentimentIntensityAnalyzer object
    sid_obj = SentimentIntensityAnalyzer()

    # Open the file and read its content
    try:
        with open(positive, 'r') as file:
            text = file.read()
    except FileNotFoundError:
        print("The specified file was not found.")
        return

    # Get the polarity scores for the text
    sentiment_dict = sid_obj.polarity_scores(text)

    # Display the sentiment scores
    print("Overall sentiment dictionary is:", sentiment_dict)
    print("Text was rated as", sentiment_dict['neg'] * 100, "% Negative")
    print("Text was rated as", sentiment_dict['neu'] * 100, "% Neutral")
    print("Text was rated as", sentiment_dict['pos'] * 100, "% Positive")

    print("Overall Sentiment is:", end=" ")
    if sentiment_dict['compound'] >= 0.05:
        print("Positive")
    elif sentiment_dict['compound'] <= -0.05:
        print("Negative")
    else:
        print("Neutral")

# Example usage
file_path = "example_text.txt"  # Replace with the actual path to your file
sentiment_scores(positive)
```

```python
from textblob import TextBlob
```

```python
# Function to analyze sentiment
def analyze_sentiment(sentence):
    blob = TextBlob(sentence)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity
    return polarity, subjectivity

# File path
file_path = neutral  # Replace with your actual file path

try:
    # Read sentences from the file
    with open(file_path, "r") as file:
        lines = file.readlines()
    
    # Analyzing each sentence in the file
    for sentence in lines:
        sentence = sentence.strip()  # Remove any extra spaces or newline characters
        if sentence:  # Skip empty lines
            polarity, subjectivity = analyze_sentiment(sentence)
            print(f"Sentence: {sentence}")
            print(f"  Polarity: {polarity:.2f}, Subjectivity: {subjectivity:.2f}\n")
except FileNotFoundError:
    print(f"File not found: {file_path}")
```
