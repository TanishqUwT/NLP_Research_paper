This is for positive sentence

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Function to calculate sentiment scores
def sentiment_scores(sentence):
    # Create a SentimentIntensityAnalyzer object
    sid_obj = SentimentIntensityAnalyzer()

    # Get the polarity scores for the input sentence
    sentiment_dict = sid_obj.polarity_scores(sentence)
    
    # Prepare the output dictionary for the sentence
    result = {
        "sentence": sentence,
        "neg": sentiment_dict['neg'] * 100,
        "neu": sentiment_dict['neu'] * 100,
        "pos": sentiment_dict['pos'] * 100,
        "compound": sentiment_dict['compound'],
        "overall_rating": (
            "Positive" if sentiment_dict['compound'] >= 0.05 else
            "Negative" if sentiment_dict['compound'] <= -0.05 else
            "Neutral"
        )
    }
    return result

# Function to analyze sentiments from a file
def analyze_file(file_path):
    results = []  # To store results for all sentences
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            sentences = file.readlines()
            print("\nAnalyzing sentiments from file:", file_path)
            for sentence in sentences:
                # Remove leading/trailing whitespace
                sentence = sentence.strip()
                if sentence:  # Ignore empty lines
                    result = sentiment_scores(sentence)
                    results.append(result)
    except FileNotFoundError:
        print("Error: File not found. Please provide a valid file path.")
        return
    except Exception as e:
        print(f"An error occurred: {e}")
        return

    # Display all results together
    print("\nSentiment Analysis Results:")
    for idx, res in enumerate(results, start=1):
        print(f"\nSentence {idx}: {res['sentence']}")
        print(f"Negative: {res['neg']}%")
        print(f"Neutral: {res['neu']}%")
        print(f"Positive: {res['pos']}%")
        print(f"Overall Rating: {res['overall_rating']}")
        print(f"Compound Score: {res['compound']}")

# Example usage
file_path = r"Z:\Programming\MyML\NLP COLLAGE\labeling the data\NLP_Research_paper\Positive.txt"   # Replace with your file path
analyze_file(file_path)
```
